```python
# 실습 과제 1. 누적합(Prefix Sum)

steps = [5231, 6120, 4890, 7005, 6500, 7200, 6800]

prefix = [0]

for s in steps:
    prefix.append(prefix[-1] + s)

for i in range(1, len(prefix)):
    print(f"{i}일 누적 : {prefix[i]}")
```

```python
# 실습 과제 2. 누적합(Prefix Sum) - 1번 응용

steps = [
    {"2026-01-01": 5231},
    {"2026-01-02": 6210},
    {"2026-01-03": 4890},
    {"2026-01-04": 7005},
    {"2026-01-05": 6500},
    {"2026-01-06": 7200},
    {"2026-01-07": 6800}
]

prefix = [0]

for day in steps:
    for date, count in day.items():
        prefix.append(prefix[-1] + count)


for i, day in enumerate(steps):
    for date, count in day.items():
        print(f"{date} 누적 : {prefix[i+1]}")
```

```python
# 실습 과제 3. 슬라이딩 윈도우

sales = [12, 15, 18, 20, 17, 25, 22, 19]

t = 3

window = sum(sales[:t])
max_sum = window

print(f"{sales[0]} + {sales[1]} + {sales[2]} = {window}")

for i in range(t, len(sales)):
    window = window - sales[i - t] + sales[i]

    print(f"{sales[i-2]} + {sales[i-1]} + {sales[i]} = {window}")

    if window > max_sum:
        max_sum = window

print(f"\n최대 매출 합 : {max_sum}")
```

```python
# 실습 과제 4. 슬라이딩 윈도우 - 3번 응용

sales = [
    {"9시": 12},
    {"10시": 15},
    {"11시": 18},
    {"12시": 20},
    {"13시": 17},
    {"14시": 25},
    {"15시": 22},
    {"16시": 19},
    {"17시": 24},
    {"18시": 21},
    {"19시": 18},
    {"20시": 16},
    {"21시": 14},
    {"22시": 11}
]

t = 3

window = 0

for i in range(t):
    for time, money in sales[i].items():
        window += money

max_sum = window

start_time = list(sales[0].keys())[0]
end_time = list(sales[t - 1].keys())[0]
max_time = f"{start_time} ~ {end_time}"

print(f"{start_time} ~ {end_time} : {window}")

for i in range(t, len(sales)):
    for time, money in sales[i - t].items():
        out_money = money
    for time, money in sales[i].items():
        in_money = money
    window = window - out_money + in_money

    start_time = list(sales[i - t + 1].keys())[0]
    end_time = list(sales[i].keys())[0]

    print(f"{start_time} ~ {end_time} : {window}")

    if window > max_sum:
        max_sum = window
        max_time = f"{start_time} ~ {end_time}"

print()
print(f"최대 매출 합 : {max_sum}")
print(f"시간대 : {max_time}")
```
