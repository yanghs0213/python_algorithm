```python
n = int(input("숫자 개수 : "))

numbers = []

for i in range(n):
    numbers.append(i)

print(numbers)
```

```python
n = int(input("크기 : "))

matrix = [] # 3 -> 3*3

for i in range(n): # for i in range(n)*range(n)
    row = []
    for j in range(n):
        row.append(0)
    matrix.append(row)

print(matrix)
```
