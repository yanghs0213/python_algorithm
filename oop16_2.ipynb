```python
class Student:
    def __init__(self, name):
        self.name = name
    def print_name(self):
        print(self.name)
    def greet(self):
        print(f"{self.name}님 hello")
```

```python
# 상속하면
class HighStudent(Student):
    def study(self):
        print("공부한다.")
```

```python
# 사용
s = HighStudent("홍길동")
s.print_name()
s.study()
s.greet()
```

```python
class Student:

    def __init__(self, name, kor, eng, math):
        self.name = name
        self.kor = kor
        self.eng = eng
        self.math = math

    def average(self):
        return (self.kor + self.eng + self.math) / 3


s = Student("홍길동", 90, 85, 100)
s2 = Student("김철수", 50, 60, 75)

print(s.average())
print(s2.average())
```
