```python
# 실습 1. 클래스 설계하기

class Product:
    # 생성자
    def __init__(self, name, price, stock):
        self.name = name # 상품명
        self.price = price # 가격
        self.stock = stock # 재고

    # 총 판매금액 반환 (가격 x 판매수량)
    def total_sales(self, quantity):
        return self.price * quantity
    
    # 상품 정보 출력
    def print_info(self):
        print("상품명 :", self.name)
        print("가격 :", f"{self.price}원")
        print("재고 :", f"{self.stock}개")
        print()

    def stock_in(self, quantity):
        self.stock += quantity

    def stock_out(self, quantity):
        if self.stock >= quantity:
            self.stock -= quantity
        else:
            print("재고가 부족합니다.")

    def change_price(self, new_price):
        if new_price >= 0:
            self.price = new_price
        else:
            print("가격은 0원보다 작을 수 없습니다.")

    # 주문
    def order(self, count):
        if self.stock >= count:
            self.stock -= count
            return self.price * count
        else:
            print("재고 부족")
            return 0

    # 주문 금액 계산
    def total_price(self, count):
        return self.price * count

    # 품절 여부
    def is_sold_out(self):
        return self.stock == 0            

# 객체 생성
p = Product("키보드", 50000, 10)

# 속성 출력
print(p.name)
print(p.price)
print(p.stock)
```

```python
# 실습 7. 딕셔너리 기반 프로그램을 클래스로 리팩터링

# Product 클래스
class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

    # 상품 정보 출력
    def print_info(self):
        print("상품명 :", self.name)
        print("가격 :", self.price)
        print("재고 :", self.stock)

    # 주문하기
    def order(self, count):
        if self.stock >= count:
            self.stock -= count
            return self.price * count
        else:
            print("재고 부족")

    # 입고하기
    def stock_in(self, count):
        self.stock += count

    # 출고하기
    def stock_out(self, count):
        if self.stock >= count:
            self.stock -= count
        else:
            print("재고 부족")

    # 가격 변경
    def change_price(self, price):
        self.price = price

    # 할인 적용
    def discount(self, rate):
        self.price = int(self.price * (1 - rate))

    # 품절 확인
    def is_sold_out(self):
        return self.stock == 0


# 딕셔너리 → Product 객체로 변경
products = [
    Product("키보드", 50000, 10),
    Product("마우스", 30000, 20),
    Product("모니터", 250000, 5)
]


# 전체 상품 조회
for p in products:
    p.print_info()
    print()


# 상품 검색
search_name = "키보드"

for p in products:
    if p.name == search_name:
        print("검색 결과")
        p.print_info()


# 주문하기
for p in products:
    if p.name == "키보드":
        print("결제금액 :", p.order(2))


# 입고하기
for p in products:
    if p.name == "마우스":
        p.stock_in(5)


# 출고하기
for p in products:
    if p.name == "모니터":
        p.stock_out(1)


# 가격 변경
for p in products:
    if p.name == "모니터":
        p.change_price(270000)


# 할인 적용 (10%)
for p in products:
    if p.name == "키보드":
        p.discount(0.1)


# 품절 상품 조회
print("품절 상품")

for p in products:
    if p.is_sold_out():
        print(p.name)


# 가장 비싼 상품 찾기
max_product = products[0]

for p in products:
    if p.price > max_product.price:
        max_product = p

print("가장 비싼 상품 :", max_product.name)


# 재고가 가장 많은 상품 찾기
max_stock_product = products[0]

for p in products:
    if p.stock > max_stock_product.stock:
        max_stock_product = p

print("재고가 가장 많은 상품 :", max_stock_product.name)


# 전체 상품 재고 수량 합계
total_stock = 0

for p in products:
    total_stock += p.stock

print("전체 재고 수량 :", total_stock)


# 전체 재고 금액 계산
total_price = 0

for p in products:
    total_price += p.price * p.stock

print("전체 재고 금액 :", total_price)
```
