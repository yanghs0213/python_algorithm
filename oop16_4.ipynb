```python
# 세일즈 목록
sales = [
    Product("키보드",50000,10),
    Product("키보드",50000,10),
    Product("마우스",30000,20),
    Product("모니터",250000,5),
    Product("노트북",1200000,3),
    Product("노트북",1200000,3),
    Product("노트북",1200000,3)
]

# Product("키보드",50000,10), 2
# Product("마우스",30000,20), 1
# Product("모니터",250000,5), 1
# Product("노트북",1200000,3), 3

# {"키보드" : 2, "마우스" : 1, "모니터" : 1, "노트북" : 3}
counts = {}
for s in sales:
    if s.name not in counts: #counts에 상품이름 키가 없다면
        counts[s.name] = 1
    else:
        counts[s.name] = counts[s.name] + 1 # counts[s.name] += 1

print(counts)

check_name_list = [] # 같은 상품이름 2번 계산되면 안되니까 계산했으면 입력
for s in sales:
    if s.name not in check_name_list:
        # count = counts[s.name]
        print(f"{s.name} : {s.total_price(counts[s.name])}")
        check_name_list.append(s.name)
```

```python
# 실습 1. 클래스 설계하기

class Product:
    # 생성자
    def __init__(self, name, price, stock):
        self.name = name # 상품명
        self.price = price # 가격
        self.stock = stock # 재고

    # 총 판매금액 반환 (가격 x 판매수량)
    def total_sales(self, quantity):
        return self.price * quantity
    
    # 상품 정보 출력
    def show_info(self):
        print("상품명:", self.name)
        print("가격:", self.price)
        print("재고:", self.stock)

# 객체 생성
p = Product("키보드", 50000, 10)

# 속성 출력
print(p.name)
print(p.price)
print(p.stock)
```

```python
# 실습 3. 객체 리스트 관리

# 객체 리스트 생성
products = [
    Product("키보드", 50000, 10),
    Product("마우스", 30000, 20),
    Product("모니터", 250000, 5),
    Product("노트북", 1200000, 3)
]

# 상품 정보 출력
for p in products:
    print("상품명:", p.name)
    print("가격:", p.price)
    print("재고:", p.stock)
    print()
# 가격이 100,000원 이상인 상품만 출력
for p in products:
    if p.price >= 100000:
        print(p.name, p.price)
# 재고가 10개 이하인 상품만 출력
for p in products:
    if p.stock <= 10:
        print(p.name, p.stock)

# 판매 목록에서 상품별 total_price 구하기
sales = [
    ("키보드", 50000, 1),
    ("마우스", 30000, 1),
    ("모니터", 250000, 1),
    ("노트북", 1200000, 1)
]

for name, price, quantity in sales:
    total_price = price * quantity
    print(name, "판매금액:", total_price)
```
