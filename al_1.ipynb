```python
# 8
try:
    num = int(input("정수를 입력하세요: "))

    if num <= 1:
        print("소수가 아닙니다")
    else:
        is_prime = True

        for i in range(2, num):
            if num % i == 0:
                is_prime = False
                break

        if is_prime:
            print("소수입니다")
        else:
            print("소수가 아닙니다")
        
except ValueError as e:
    print("숫자만 입력하세요")
```

```python
# 9
count = 0

for num in range(3, 101):
    is_prime = True

    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        count += 1

print(count)
```

```python
# 11
n = 10

for i in range(n):
    for j in range(n):
        if i == j or i + j == n - 1:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```
