```python
import re

# ^(캐럿 caret) not
text = "a7bc123"
result = re.findall(r"[^0-9]", text) # 0~9문자가 아니면 반환되는 리스트에 추가
print(result)
```

```python
import re

text = "7bc123"
result = re.findall(r"^[0-9]", text) # 0~9문자가 아니면 반환되는 리스트에 추가
print(result)
```

```python
import re

text = "cat ct cot cut coot"
result = re.findall(r"c.t", text) # "c_t" 이런 패턴의 문자(c로 시작, t로 끝나는 문자열길이 3) 찾기
print(result)
```

```python
import re

# findall함수에서는 ^패턴을 사용해도 의미있는 반환값을 얻을 수 없다(^패턴은 전체 문자열을 하나로 다루기 때문)
text = "Python Python python"
print(re.findall(r"^Py", text)) # ["Py"]

text = "python Python python"
print(re.findall(r"^Py", text)) # [] 

print("-"*10)
texts = ["Python Python python", "python Python python"]
for t in texts:
    print(re.findall(r"^Py", t))
```

```python
import re

# findall함수에서는 ^패턴을 사용해도 의미있는 반환값을 얻을 수 없다(^패턴은 전체 문자열을 하나로 다루기 때문)
text = "Python Python python"
print(re.findall(r"(Py)", text)) # ["Py"]

# \d{3}-\d{4}-\{4}
# ^(010)-(\d{4})-(\{4})$
```

```python
import re

text = "Python Python python"
print(re.search(r"^Py", text)) # <re.Match object; span=(0, 2), match='Py'> span은 Py문자의 위치 0에서 2-1에 위치
print(re.search(r"^py", text)) # None
```

```python
import re

text = "Python Python python"
print(re.match(r"Py", text)) # <re.Match object; span=(0, 2), match='Py'> span은 Py문자의 위치 0에서 2-1에 위치
print(re.match(r"py", text)) # None
print(re.match(r"Py", text).group())
```

```python
import re

text = "Pyyyyython"

# 시작이 Py 찾기
print(re.search(r"^Py", text)) # <re.Match object; span=(0, 2), match='Py'>
# 시작이 Pyyyyy.... 찾기
print(re.search(r"^Py*", text)) # <re.Match object; span=(0, 6), match='Pyyyyy'>
```

```python
import re

texts = ["3", "123", "", "abc", "12a"]

for t in texts:
    print(t, "->", bool(re.fullmatch(r"\d+", t))) # \d* : \d+ : 숫자가 1개 이상 있어야 한다.
```

```python
import re

texts = ["3", "123", "", "abc", "12a"]

for t in texts:
    print(t, "->", bool(re.fullmatch(r"\d*", t))) # \d* : \d+ : 숫자가 1개 이상 있어야 한다.
```

```python
import re

words = [
    "ae",
    "ac",
    "abc",
    "abbcd",
    "abbbcde",
    "acde",
    "axc"
]

# ab?c : b문자가 0개 or 1개 패턴

for word in words:
    print(re.match(r"ab?c", word))
# None
# <re.Match object; span=(0, 3), match='abc'>
# None
# None
# <re.Match object; span=(0, 2), match='ac'>
```
