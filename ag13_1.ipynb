```python
scores = [10, 20, 3, 40, 5, 60, 8, 55]

k = 3 # window size

# window 시작 포인트(인덱스 0)의 합을 초기값으로 설정
window = sum(scores[0:k]) # scores[0:k]

# 초기값에서 1칸을 이동 -> k까지
# window + scores[k] - scores[k-k]
# window + scores[k+1] - scores[k+1 - k]

print(window) # 60

for i in range(k, len(scores)): # range(3, 6) 3칸이동 / 원도우크기 부터 1칸식 이동하며 윈도우안의 합을 연산
    window = window - scores[i-k] + scores[i]
    # window + scores[3] - scores[3-3] -> 90
    # window + scores[4] - scores[4-3] -> 120
    # window + scores[5] - scores[5-3] -> 150
    print(window)
```

```python
scores = [10, 20, 3, 40, 5, 60, 8, 55, 44]

k = 3 # window size

# window 시작 포인트(인덱스 0)의 합을 초기값으로 설정
window = sum(scores[0:k]) # scores[0:k]

totals = [window]
avgs = [window/k]
max_val = {"구간":"0~2","합":window}
min_val = {"구간":"0~2","합":window}

print(window) # 60

for i in range(k, len(scores)): # range(3, 6) 3칸이동 / 원도우크기 부터 1칸식 이동하며 윈도우안의 합을 연산
    window = window + scores[i] - scores[i-k]
    totals.append(window)
    avgs.append(window/k)

    # section = f"{i-k+1} ~ {i}"
    section = f"{i-(k-1)} ~ {i}"

    if window > max_val["합"]:
        max_val = {"구간":section,"합":window}
        print("max_val:", max_val)
    if window < min_val["합"]:
        min_val = {"구간":section,"합":window}

    #print(window)

print(totals)
print(avgs)
print(max_val)
print(min_val)
```

```python
# 입력값
fruit = {
    "color": "Red",
    "size": "Large" # Small
}

fruit_tree = {
    "question": ("color", "Red"),

    "yes": {
        "question": ("size", "Large"),
        "yes": "사과 (Apple)",
        "no": "체리 (Cherry)"
    },

    "no": {
        "question": ("color", "Purple"),
        "yes": "포도 (Grape)",
        "no": "모르는 과일"
    }
}

node = fruit_tree # node변수의 초기값은 dict -> 무한루프 -> node변수값이 dict 아닐때까지

# 트리를 탐색하는 반복문
# 2번 반복

# 1번째 node = "question": ("color", "Red"),

#    "yes": {
#        "question": ("size", "Large"),
#        "yes": "사과 (Apple)",
#        "no": "체리 (Cherry)"
#    },

#    "no": {
#        "question": ("color", "Purple"),
#        "yes": "포도 (Grape)",
#        "no": "모르는 과일"
#    }
#}

# 2번째 node = while isinstance(node, dict): → True

# 3번째 node = while isinstance(node, dict): → False

while isinstance(node, dict): # node객체가 dict 이라면 True 아니면 False

    # dict이라면 "question"키값을 찾아 질문은 key, 정답은 value에 저장 
    key, value = node["question"] 
   
    # 입력데이터의 key값과 비교하여 정답이면 "yes"키값을 node에 대입, 아니면 "no"키값을 node에 대입
    # 이렇게 대입된 node값이 dict 타입이면 다시 반복문 진행... 
    # 이렇게 node에 dict타입이 아닌 타입(위 트리에선 문자열)이 올때 까지 반복문은 계속
    if fruit[key] == value:
        node = node["yes"]
    else:
        node = node["no"]

# 반복문을 빠져 나오면 마지막 node값을 출력
print(node)
```
