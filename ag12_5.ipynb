```python
from collections import deque

dq = deque()
dq.append(10)
dq.append(20)
dq.append(30)

print(dq)
```

```python
dq = deque()
dq.append(20)
dq.append(30)
dq.appendleft(10) # 0번째 위치에 삽입

print(dq)

x = dq.pop() # 스텍처럼.....

print(x)
print(dq)

print(dq.popleft()) # 큐처럼.....
print(dq)
```

```python
# 스텍이 필요 -> 효율성 Deque X -> list 사용
# 큐 필요 -> 효율성 O -> Deque 사용
# 덱 필요 -> Deque 사용
# 자료구조의 회전 -> 시계방향? 반시계? 양방향 삽입/삭제 가능 -> 덱
```

```python
dq = deque([1, 2, 3, 4, 5])
dq.rotate(1) # 우측으로 한칸 이동(회전)

print(dq) # deque([5, 1, 2, 3, 4])

dq1 = deque([1, 2, 3, 4, 5])
x = dq1.pop()
dq1.appendleft(x)
print(dq1)
```

```python
# 기본 모듈안에 덱을 사용하는 이유는 덱이라는 자료구조 때문
# 덱 편리성 메서드를 많이 제공(큐, 스텍, 회전, 최대개수, ...)
```

```python
history = deque(maxlen=5) # maxlen=n n개만 저장

history.append("google")
history.append("naver")
history.append("youtube")
history.append("github")
history.append("chatgpt")

print(history)

history.append("python")

print(history)
```

```python
from collections import deque

cards = deque([1, 2, 3, 4, 5]) # 처음 받은 카드 5장

while len(cards) > 1:
    print("버림 :", cards.popleft()) # 1번 카드를 버림
    cards.rotate(-1)
    # cards.append(cards.popleft())
    print("남은 카드 :", cards)

print("마지막 카드 :", cards[0])
```
